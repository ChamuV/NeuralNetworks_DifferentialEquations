```python
# The heat equation is a PDE that describes how heat diffuses through a given region over time.
# It is simply stated as:
#     ∂u/∂t = D ∇²u

# Where 
# u(x, t): temperature at position x and time t
# D > 0: diffusion coefficient
# domain: x ∈ [0, L], t > 0

# We also provide 
# Initial condition: u(x, 0) = u0(x) for x ∈ [0, L]
# Boundary conditions: u(0, t) = u(L, t) = 0 for t > 0
```
